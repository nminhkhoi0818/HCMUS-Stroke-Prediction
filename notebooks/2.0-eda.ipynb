### 5. What is the current data type of each column? Are there columns having inappropriate data types?

```python
dtypes = raw_df.dtypes
dtypes
```

We observe that the column `age` has `float64` type. However, it actually preresents an integer which is the age of patient. Thus, it should be converted to `int64` type. In term of `hypertension`, `heart_disease` and `stroke`, they show that the status of disease. So that, it can be converted to categorical type.

```python
raw_df['age'] = raw_df['age'].astype(np.int64)
raw_df['hypertension'] = raw_df['hypertension'].astype(str)
raw_df['heart_disease'] = raw_df['heart_disease'].astype(str)
raw_df.dtypes
```

### 6. With each numerical column, how are values distributed?

```python
num_col_info_df = raw_df.select_dtypes(exclude='object')

def missing_ratio(s):
    return (s.isna().mean() * 100).round(1)

def median(df):
    return (df.quantile(0.5)).round(1)

def lower_quantile(df):
    return (df.quantile(0.25)).round(1)

def upper_quantile(df):
    return (df.quantile(0.75)).round(1)
    
num_col_info_df = num_col_info_df.agg([missing_ratio, 'min', lower_quantile, median, upper_quantile, 'max'])
num_col_info_df
```

**Are they abnormal?**
- All of them are normal.

### 7. Preprocess missing values

```python
def filling_missing_value(df: pd.DataFrame) -> pd.DataFrame:
    if (df is None):
        print(f"[LOG] Invalid.")
        raise ValueError
    
    # Create a copy of the dataframe to avoid changing the original
    df_cp = df.copy()
    
    # Get the list of columns in the dataframe
    cols_list = ['id', 'age', 'avg_glucose_level', 'bmi']
    
    # Replace all None values with NaN, fillna only works on nans
    df_cp.fillna(value=np.nan, inplace=True)
    
    # Replace all NaN values with the mean of the column values
    for col in cols_list:
        df_cp[col].fillna((df_cp[col].median()), inplace=True)
        
    return df_cp
```

```python
raw_df = filling_missing_value(df = raw_df)
raw_df.head()
```

```python
raw_df.select_dtypes(exclude='object').agg([missing_ratio, "min", lower_quantile, median, upper_quantile, "max"])
```

### 8. With each categorical column, how are values distributed?

```python
pd.set_option('display.max_colwidth', 100) # For clearly
pd.set_option('display.max_columns', None) # For clearly
```

```python
cat_col_info_df = raw_df.select_dtypes(include='object')

def missing_ratio(s):
    return (s.isna().mean() * 100).round(1)

def num_values(s):
    s = s.str.split(';')
    s = s.explode()
    return len(s.value_counts())

def value_ratios(s):
    s = s.str.split(';')
    s = s.explode()
    totalCount = (~s.isna()).sum()
    return ((s.value_counts()/totalCount*100).round(1)).to_dict()

cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df
```

**Are they abnormal?**
- All of them are normal
